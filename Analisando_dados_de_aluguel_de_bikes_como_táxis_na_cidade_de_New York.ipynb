## Transformando a coluna da em Data

```python
# Agora vamos transformar a coluna data no tipo Data
df = pd.read_csv('dataframe_saved_v1.csv', parse_dates = ['Data'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
cols = ['Data', 'Distancia', 'Tempo']
df.columns = cols
df.head()
```

## Definindo a coluna Data como index

```python
df.set_index('Data', inplace = True)
```

```python
df.head()
```

```python
df.plot()
```

## Transfornado a coluna de duração no tempo em segundos

```python
def calcula_total_segundos(time):
    if time is np.nan:
        return np.nan
    hrs, mins, seconds = str(time).split(':')
    seconds = int(seconds) + 60 * int(mins) + 60 * 60 * int(hrs)
    return seconds
```

## Adicionando uma coluna Segundos no DataFrame

```python
df['Segundos'] = df.Tempo.map(calcula_total_segundos)
```

```python
df.head(10)
```

```python
df.describe()
```

## Preenchendo valores NAN com 0

```python
df.fillna(0).describe()
```

## Adicionando coluna Minutos no DataSet

```python
df['Minutos'] = df['Segundos'].map(lambda x: x / 60)
```

```python
df.fillna(0).describe()
```

## Verificando se há uma relação entre as varíaveis com Scatter Plot

A grafico abaixo mostra que a medida que a distância aumenta leva-se mais tempo para a corrida

```python
df.plot(x = 'Distancia', y = 'Minutos', kind = 'scatter')
```

```python
# Criando tabela de relação
df.corr()
```

## Criando bloxplot para verificar Outliers

```python
df.boxplot('Minutos', return_type = 'axes')
```

## Criando variável Minutos por KM

```python
df['Min_Por_km'] = df['Minutos'] / df['Distancia']
```

```python
df.fillna(0).describe()
```

## Criando um Histograma para mostrar a frequencia da variável

```python
df.hist('Min_Por_km')
```

```python
df.hist('Min_Por_km', bins = 20)
```

```python
df.hist('Min_Por_km', bins = 20, figsize = (10, 8))
plt.xlim((5, 11))
plt.ylim((0, 12))
plt.title("Histograma Minutos Por KM")
plt.grid(True)
```
